```python
# w = 640
# h = 640

final_df = []
label_path = '/home/venkat/Venkat/Mosquito Detection/yolov5/runs/detect/exp21/labels'

for item in image_file_names:
    original_img_dir = '/home/venkat/Venkat/Abdomen Classification/M1_Database'
    original_im_path = original_img_dir + '/' + item[:-4] + '.jpg'
    
    if not os.path.exists(original_im_path):
        original_im_path = '/home/venkat/Venkat/Abdomen Classification/M2_Database' + '/' + item[:-4] + '.jpg'
    
#     im = Image.open(original_im_path)
    im = cv2.imread(original_im_path)
#     w, h = im.size
    h, w, channels = im.shape 
    
    
    filelabel = open(os.path.join(label_path, item), "r")
    lines = filelabel.read().split('\n')
    obj = lines[:len(lines) - 1]
    row = []
    max_conf = 0
    for i in range(0, int(len(obj))):
        
        splited = obj[i].split(' ')
        
        try:
            img_name = item[:-4]
            img_name = img_name + '.jpg'
            x1 = float(splited[1])
            y1 = float(splited[2])
            w1 = float(splited[3])
            h1 = float(splited[4])
            
            cur_conf = float(splited[5])
            if cur_conf > max_conf:
                xmin = int((x1*w) - (w1*w)/2.0)
                ymin = int((y1*h) - (h1*h)/2.0)
                xmax = int((x1*w) + (w1*w)/2.0)
                ymax = int((y1*h) + (h1*h)/2.0)
                x1_m = float(splited[1])
                y1_m = float(splited[2])
                w1_m = float(splited[3])
                h1_m = float(splited[4])
#                 xmin = int((x1*w))
#                 ymin = int((y1*h))
#                 xmax = int((w1*w))
#                 ymax = int((h1*h))
                max_conf = cur_conf
        except exception as e:
            print(e)
            print("File Not in YOLO format")
            
            
    row.append(img_name)
    row.append(xmin)
    row.append(ymin)
    row.append(xmax)
    row.append(ymax)
    row.append(splited[0])
    row.append(max_conf)
    row.append(x1_m)
    row.append(y1_m)
    row.append(w1_m)
    row.append(h1_m)
    
    final_df.append(row)
```

```python
df = pd.DataFrame(final_df, columns = ['image', 'xmin', 'ymin', 'xmax', 'ymax', 'label', 'max_conf', 'x1_m', 'y1_m', 'w1_m', 'h1_m'])
```

```python
from PIL import Image
import os

def crop(im):
    """
    input the image, return the cropped part of the image
    : im: image to crop
    : returns cropped image
    """
    # Size of the image in pixels (size of original image)
    width, height = im.size

    if width > height:
        diff = width - height
        top = 0
        bottom = height
        left = diff//2
        right = width - (diff//2)
        crop_img = im.crop((left, top, right, bottom))

    elif height > width:
        diff = height - width
        top = diff//2
        bottom = height - (diff//2)
        left = 0
        right = width
        crop_img = im.crop((left, top, right, bottom))
    else:
        crop_img = im
    
    return crop_img

def pad(im, color):
    """
    Pads the image to make it a square"
    : im: image to pad
    : color: color to pad the image
    : returns padded image
    """
    width, height = im.size

    # if widht and height are same, padding is not required.
    if width == height:
        print("I reached here")
        pad_img = im
    
    elif width > height:
        pad_img = Image.new(im.mode, (width, width), color)
        pad_img.paste(im, (0, (width - height) // 2))
    
    else:
        pad_img = Image.new(im.mode, (height, height), color)
        pad_img.paste(im, ((height - width)//2, 0))
    
    return pad_img
```
